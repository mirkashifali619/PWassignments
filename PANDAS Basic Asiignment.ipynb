````python
#Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
#ANS:
#In pandas, a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure. It is one of the most commonly used data structures in pandas and is similar to a spreadsheet or SQL table. DataFrames are used to store and manipulate data in a structured way, making them a powerful tool for data analysis and manipulation.

#Here are some key characteristics of a DataFrame:

#1. Two-Dimensional: A DataFrame is organized as a two-dimensional table with rows and columns. Rows represent individual observations or records, while columns represent different variables or attributes.

#2. Size-Mutable: You can add or remove rows and columns from a DataFrame after it's created. This makes it flexible for data manipulation.

#3. Heterogeneous Data: DataFrames can store data of different types (e.g., integers, floats, strings) in different columns, making them suitable for handling real-world datasets with mixed data types.

#On the other hand, a pandas Series is a one-dimensional array-like object that can hold data of any type. A Series is essentially a single column from a DataFrame. Each element in a Series has an associated label called an index. Series can be thought of as a special case of a DataFrame where you have only one column.

#Let's illustrate the difference between a DataFrame and a Series with an example:

#```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)

# df is now a DataFrame with three columns: 'Name', 'Age', and 'City'
print(df)

# Creating a Series from the 'Age' column of the DataFrame
age_series = df['Age']

# age_series is a Series containing the ages of individuals
print(age_series)

# Accessing a specific element in the DataFrame
print(df.loc[0, 'Name'])  # Accessing the name of the first person

# Accessing a specific element in the Series
print(age_series[1])  # Accessing the age of the second person
#```

#In this example, `df` is a DataFrame containing information about individuals, including their names, ages, and cities. `age_series` is a Series created from the 'Age' column of the DataFrame. While both DataFrame and Series can store and manipulate data, the key difference is in their dimensionality and how they are accessed. DataFrames are two-dimensional, and you can access data by both row and column labels, while Series are one-dimensional and have a single column of data with associated index labels.
````

````python
#Q7. Create a DataFrame using multiple Series. Explain with an example.
#ANS:
#You can create a DataFrame using multiple Series in pandas by combining these Series into a dictionary and then using the `pd.DataFrame()` constructor. Each Series will become a column in the resulting DataFrame, and they should have the same length to align properly. Here's an example:

#```python
import pandas as pd

# Creating multiple Series
name_series = pd.Series(['Alice', 'Bob', 'Charlie'])
age_series = pd.Series([25, 30, 35])
city_series = pd.Series(['New York', 'San Francisco', 'Los Angeles'])

# Combining the Series into a dictionary
data = {'Name': name_series, 'Age': age_series, 'City': city_series}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(data)

# Displaying the resulting DataFrame
print(df)
#```

#In this example, we first create three Series (`name_series`, `age_series`, and `city_series`) with information about individuals' names, ages, and cities. Then, we combine these Series into a dictionary called `data`, where each Series becomes a column in the DataFrame. Finally, we use the `pd.DataFrame(data)` constructor to create the DataFrame `df`. The resulting DataFrame will look like this:

#```
#      Name  Age           City
#0    Alice   25       New York
#1      Bob   30  San Francisco
###2  Charlie   35    Los Angeles
#```

#Now, you have a DataFrame with columns 'Name,' 'Age,' and 'City,' where each column corresponds to one of the Series you initially created.
````
